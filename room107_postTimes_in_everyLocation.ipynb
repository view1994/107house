```python
import pymongo
import charts
```

```python
client=pymongo.MongoClient('localhost',27017)
room107=client['room107']
room_info=room107['room_info']
```

```python
for i in room_info.find().limit(300):
    print(i['locations'][0].split('&nbsp'))
```

```python
for i in room_info.find():
    if i['locations']:
        locations=i['locations'][0].split('&nbsp')[1:]+[i['locations'][1]if len(i['locations'])==2 else '楼层不明']
    else:
        locations=['unknown']
    room_info.update({'_id':i['_id']},{'$set':{'locations':locations}})
```

```python
for i in room_info.find().limit(300):
    print(i['locations'])
```

```python
location_list=[]
for i in room_info.find():
    location_list.append(i['locations'][0])
location_index=list(set(location_list))
print(location_index,len(location_index))
```

```python
post_times=[]
for i in location_index:
    post_times.append(location_list.count(i))
print(post_times)
```

```python
series=[]
for i,n in zip(location_index,post_times):
    series.append({'name':i,'data':[n],'type':'column'})
charts.plot(series,show='inline',options=dict(title=dict(text='post times in every location')))
```
