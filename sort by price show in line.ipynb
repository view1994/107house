```python
def get_date_in(date,date1,date2):    
    if (date[0]>= date1[0]) & (date[0]<=date2[0]) & (date[1]>=date1[1]) & (date[1]<=date2[1]) & (date[2]>=date1[2]) & (date[2]<=date2[2]):
        return True
    else:
        return False

#get_date_in([2017,11,15],[2017,12,1],[2017,12,31])
```

```python
location_list=[]
price_list=[]
for i in room_info.find({},{'date':1,'monthPrice':1,'locations':1,'_id':0}):
    if  get_date_in(i['date'],[2017,12,1],[2017,12,31]):
        location_list.append(i['locations'][0])
        price_list.append(i['monthPrice'])
location=set(location_list)
price=set(price_list)

print(location,price,len(location),len(price),len(location_list),len(price_list))
```

```python
def sort_by(p,l,count):
    if p in range(0, 1000):
        count[l][0] += 1
    elif p in range(1000, 2000):
        count[l][1] += 1
    elif p in range(2000, 3000):
        count[l][2] += 1
    elif p in range(3000, 4000):
        count[l][3] += 1
    elif p in range(4000, 5000):
        count[l][4] += 1
    elif p in range(5000, 6000):
        count[l][5] += 1
    elif p in range(6000, 7000):
        count[l][6] += 1
    elif p in range(7000, 8000):
        count[l][7] += 1
    elif p in range(8000, 9000):
        count[l][8] += 1
    else:
        count[l][9] += 1
    return count
count={}
location=[]
for i in room_info.find({},{'date':1,'monthPrice':1,'locations':1,'_id':0}):
    if get_date_in(i['date'],[2017,12,1],[2017,12,31]):
        if not i['locations'][0] in location:
            location.append(i['locations'][0])
            count[i['locations'][0]] = [0] * 10
        sort_by(i['monthPrice'], i['locations'][0], count)
    else:
        print ('date over range: ',i['date'])
print (count,str(len(count)))
#charts.plot(series,show='inline',options=dict(title=dict(text='post times in every location')))
```

```python
series=[]
price_range=['RMB_1K','RMB_2K','RMB_3K','RMB_4K','RMB_5K','RMB_6K','RMB_7K','RMB_8K','RMB_9K','RMB_9K+']
for i in count:
    series.append({'name':i,'data':count[i],'type':'line'})
print(series,str(len(series)))
```

```python
options={
    'charts'  :{'zoomType':'xy'},
    'title':{'text':'各价格区间发帖量'},
    'subtitle':{'text':'可视化统计图表'},
    'xAxis'   :{'categories':price_range},
    'yAxis'   :{'title': {'text':'数量'}}
}
charts.plot(series,options=options,show='inline')
```
